# while loops

```python
x = 0

while x < 10:
    print 'x is currently: ',x
    x += 1
```

```python
x = 0

while x < 10:
    print 'x is currently: ',x
    x += 1
else:
    print 'All Done!'
```

```python
# Break, Continue and Pass
```

```python
x = 0

while x < 10:
    print 'x is currently: ',x
    print ' x is still less than 10, adding 1 to x'
    x += 1
    
    if x == 3:
        print ' Hey x equals 3!'
    else:
        print 'continuing....'
        continue
```

```python
x = 0

while x < 10:
    print 'x is currently: ',x
    print ' x is still less than 10, adding 1 to x'
    x += 1
    
    if x == 3:
        print ' Hey x equals 3!'
        break
    else:
        print 'continuing....'
        continue
```
